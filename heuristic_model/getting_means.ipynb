```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/covtype/covtype.data.gz', names=range(0, 55))
```

```python
# Model will be working based on means of Elevation, Slope and Horizontal_Distance_To_Roadways variables

df = df.rename(columns={0: "elev", 2: 'slope', 5: "road_distance", 54: 'label'})
df = df[['elev', 'slope', 'road_distance', 'label']]
df
```

```python
means_tuples = []

for label in range(1, 8):
    working_df = df[df['label'] == label]
    
    means_tuples.append((working_df['elev'].mean(),
                         working_df['slope'].mean(),
                         working_df['road_distance'].mean()))
```

```python
df_means = pd.DataFrame(means_tuples)
df_means.to_csv('means_heuristic_model.csv')
```
